## 2.1 크롤링 전 옵션 설정


```python
# input keywords
# search_words = crawling_dog

search_words = 'brittanydog'
# replace "," to ""
if "," in search_words :
    search_words = search_words.replace(',', '')

search_list = search_words

# debug
print(search_list)
search_list = list(search_list)
```

```python
# input max_limits
max_limits = int(input("max crawling count : "))


# input image_save dir
save_dir = crawling_dir
save_dir
```

```python
# default size
size_opt = '>320x320' # medium, large, icon, >[]x[], =[]x[] ([] is an integer) e.g ">640x480" (x must be small letter) 800X600

# type
type_opt = 'photo' # photo, face, clipart, linedrawing, animated

# date
# date_opt = 'pastyear' # pastday, pastweek, or a tuple of dates. e.g ((2016, 1, 1), (2017, 1, 1)) or ((2016, 1, 1), None) *None means until today

# color
# color_opt = "color" #“blackandwhite”, “red”, “orange”, “yellow”, “green”, “teal”, “blue”, “purple”, “pink”,“white”, “gray”, “black”, “brown”

# default license
# license_opt = 'commercial,modify' # noncommercial, commercial, noncommercial,modify, commercial,modify
# 개인 사용 목적일 경우 영리 이미지 사용가능
# license_opt = 'None'
```

```python
# for img_dir in search_list:
for img_dir in ['brittanydog']:

    print(img_dir)
    if not os.path.exists(os.path.join(save_dir,img_dir) ):
        os.mkdir(os.path.join(save_dir,img_dir))
        print(os.path.join(save_dir,img_dir))
    
    google_crawler = GoogleImageCrawler( 
                feeder_threads = 1,
                parser_threads = 2, 
                downloader_threads = 4,
                storage={ 'root_dir' : save_dir + "/" + img_dir })
    filters = dict(
            #size = size_opt,
#             type = type_opt,
            #license = license_opt
            )
    google_crawler.session.verify = False
    google_crawler.crawl(keyword = img_dir , 
                         filters = filters, 
                         max_num = max_limits, 
                         file_idx_offset = 0   #'auto'
                        )
    
    # debug
    print("keyword {} {} images have been saved.".format( img_dir, max_limits))
```
