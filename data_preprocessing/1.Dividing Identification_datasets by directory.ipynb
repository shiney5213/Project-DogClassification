# 1. label 만들기

## 1.1. 파일 불러오기: df_labels

```python
df_labels =pd.read_csv('../data/dog-breed-identification/labels.csv')
test_df=pd.read_csv('../data/dog-breed-identification/sample_submission.csv')
len(df_labels)
```

```python
df_labels.head()
```

## 1.2. id, beed dict로 저장 ->  labels_dict

```python
labels_breed = df_labels['breed'].values
labels_id = df_labels['id'].values
```

```python
labels_dict  = {}
for key, value in zip(labels_id, labels_breed):
    value = value.replace('_','').replace('-','')
    labels_dict[key]= value
```

```python
len(labels_dict.keys())
```

```python
labels_dict
```

## 1.3. image file name list : img_name_list

### img full path: img_path

```python
root_dir = '../data/dog-breed-identification/train/'
```

```python
img_name_list = os.listdir(root_dir)
len(img_name_list)
```

```python
img_name_list[0]
```

```python
# 이미지별 path
img_path = [os.path.join(root_dir +i)  for i in img_name_list ]
img_path
```

```python
# 파일이름 list
img_name_list = []
for i in img_path:
    img_name_list.append(os.path.split(i)[-1].replace('.jpg', ''))
```

```python
len(img_name_list)
```

```python
img_name_list[0:5]
```

## 1.4. 이미지 label완성: labels

```python
labels = []
for i in img_name_list:
    for key, value in labels_dict.items():
        if i == key:
            value = value.replace('_','').replace('-','')
            labels.append(value)
```

# 2. 이미지 저장

## 2.1. 폴더만들기

```python
root_dir = '../data/dog-breed-identification/dog_identification_120_dir/'
os.path.isdir(root_dir)
```

```python
for value in  labels_dict.values() :
    try:
        new_dir = os.path.join(root_dir, value)
        if not(os.path.isdir(new_dir)):
            os.makedirs(new_dir)
    except OSError as e:
        if e.errno != errno.EEXIST:
            print("Failed to create directory!!!!!")
            raise
```

## 2.2. 폴더 별 사진 저장

```python
for key, value, path in zip(labels_dict.keys(), labels_dict.values(), img_path):
    img = cv2.imread(path, cv2.IMREAD_COLOR)
    cv2.imwrite(os.path.join(root_dir, value, os.path.basename(path)), img)
```
